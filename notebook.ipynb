## Inferência estatística

### População e amostra
### A população são todos os ítens de interesse (pessoas, produtos, negócios, etc).
### Uma amostra é um subconjunto da população
### Uma observação é um ítem unico em uma amostra
### O censo é uma amostra que consiste na população inteira

### Uma estatística amostral é usada para estimar um parâmetro da população
### Geralmente usada devido a impossibilidade de coletar o censo(dados da população inteira) para obter o parâmetro exato
### População -> Amostra -> Estatística -> Inferência -> Parâmetro

## Distribuição amostral:
### Teoria assume que as amostras são aleatórias e representativas.
### A estatística da amostra é uma variável aleatória já que o valor da estatística da amostra varia de amostra para amostra. As estatísticas da amostra formam um padrão no longo prazo.
## Parâmetros e estatísticas:
### Existe só uma população. O parâmetro é uma constatnte cujo valor pode ser desconhecido
### Uma estatística é uma variável aleatória cujo valor depende da amostra aleatória

### População: A variável aleatória X é uma característica da população. Ela tem média (valor esperado) e desvio padrão(SD).

### Amostras: tem um erro padrão(se)

## Teorema Central do Limite:
### Afirma que quando o tamanho da amostra aumenta, a distribuição amostral da sua média aproxima-se cada vez mais de uma distribuição normal.
### A distribuição amostral é normal se:
### - As amostras foram retiradas de uma população com distribuição normal
### - Teorema Central do Limite -> Se o tamanho da amostra é n>=30

### Regressão Linear

### Método para comparação entre duas ou mais variáveis
### Uma variável dependente (variável de resposta), é influenciada por uma ou mais variávels independentes (variáveis explicativas)
### Usamos um modelo de regressão para prever os valores da variável dependente (y só existe uma) dado as variáveis independentes (x, pode haver mais de uma, com uma chama-se regressão linear simples, mais de uma = regressão linear multipla)
